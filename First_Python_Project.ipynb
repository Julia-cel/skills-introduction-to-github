```python
employment_type_map = {
    'FT': 'Full-time',
    'CT': 'Contract',
    'PT': 'Part-time',
    'FL': 'Freelance'
}

remote_ratio_map = {
    0: ' Presencial work',
    50: 'Partially remote',
    100: 'Fully remote'
}

company_size_map = {
    'S': 'Small',
    'M': 'Medium',
    'L': 'Large'
}

df['employment_type'] = df['employment_type'].replace(employment_type_map)
df['remote_ratio'] = df['remote_ratio'].replace(remote_ratio_map)
df['company_size'] = df['company_size'].replace(company_size_map)

print("Columns 'employment_type', 'remote_ratio', and 'company_size' updated with descriptive names.")

print("\nUpdated employment_type value counts:")
display(df['employment_type'].value_counts())

print("\nUpdated remote_ratio value counts:")
display(df['remote_ratio'].value_counts())

print("\nUpdated company_size value counts:")
display(df['company_size'].value_counts())
```

```python
##Creating data frame to simulate how replace nulls values for the average of values

import numpy as np
df_salary = pd.DataFrame({'nome':["Ana","Marcos","Paulo","Mel","Bob"],
                            'salary': [4000, np.nan, 3500, np.nan, 5000]
})
##Calcular a media:
df_salary['MeanSalary']= df_salary['salary'].fillna(df_salary['salary'].mean().round(2))
df_salary

##Replacing null for the median(mediana) if the dataset there's a lot of outliers that biases the data
df_salary['MedianSalary']= df_salary['salary'].fillna(df_salary['salary'].median())
df_salary
```

```python
##Creating second dexample data frame
df_temperature = pd.DataFrame({'Days':["Monday", "Tuesday","Wednesday", "Thursday","Friday"],
                               'Temp': [30,np.nan,np.nan,28,27]})
##Replacing with fffil= forward fill -> fill the null with the previous value
df_temperature['FilledNa']= df_temperature['Temp'].ffill()
df_temperature
```

```python
## Testing the opposite function of forward fill(ffill) -> Back fill(bfill)=replace with the values after NA
df_temperature['Bfill']= df_temperature['Temp'].bfill()
df_temperature
```

```python
df_cities = pd.DataFrame({
    'Names':['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'Cities':[ 'São Paulo', np.nan, 'Curitiba', np.nan, 'Salvador']
})
df_cities['FillCities']=df_cities['Cities'].fillna("Não informado")
df_cities
```

```python
## Coming back to the df to eliminate the nulls
df_clean = df.dropna()
df_clean.isnull().sum()
```

```python
df_clean.info()
```

```python
##Correcting data type of Year

df_clean=df_clean.assign(work_year= df_clean['work_year'].astype("int64"))
subex= {   'SE': 'Senior',
    'MI': 'Mid-level',
    'EN': 'Entry-level',
    'EX': 'Executive-level'}
df_clean['experience_level']= df_clean['experience_level'].replace(subex)
df_clean
```

```python
df_clean['experience_level'].value_counts().plot(kind='bar', title="Experience Level Distribution")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(data=df_clean,x="experience_level", y="salary_in_usd")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_clean,x="experience_level", y="salary_in_usd")
plt.title("Average salary per experience level")
plt.xlabel("Experience Level")
plt.ylabel(" Average annual Salary in USD")
plt.show()
```

```python
sortedSalary = df_clean.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_clean,x="experience_level", y="salary_in_usd", order =sortedSalary)
plt.title("Average salary per experience level")
plt.xlabel("Experience Level")
plt.ylabel(" Average annual Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_clean["salary_in_usd"], bins=50, kde= True)##Kde is a line to emphasize the tendency of the distribution
plt.title("Distribution of annual salaries ")
plt.xlabel("Annual Salary")
plt.ylabel("Frequency")
plt.show()

##Histogram to show the distribution of values by 50 -> this distribution show a concentration to low values
## -> different from a normal distribution when the values spread through the middle
```

```python
## Median is the line in the rectangle and the points are all outlines
plt.figure(figsize=(8,5))
sns.boxplot(x=df_clean['salary_in_usd'])
plt.title("BoxPlot salaries ")
plt.xlabel("Annual Salary")
```

```python
SortExperienceLevel= ['Entry-level', 'Mid-level', 'Senior', 'Executive-level']
plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level',y='salary_in_usd',data=df_clean, order= SortExperienceLevel )
plt.title("BoxPlot salaries by experience level ")
plt.xlabel("Executive Level")
```

```python
## Add palette and hue to colors by categories
SortExperienceLevel= ['Entry-level', 'Mid-level', 'Senior', 'Executive-level']
plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level',y='salary_in_usd',data=df_clean, order= SortExperienceLevel, palette= 'Set2',hue='experience_level' )
plt.title("BoxPlot salaries by experience level ")
plt.xlabel("Executive Level")
```

```python
import pycountry

def get_country_name(country_code):
    try:
        return pycountry.countries.get(alpha_2=country_code).name
    except:
        return country_code # Return the code if not found

employee_residence_counts = df_clean["employee_residence"].value_counts().reset_index()
employee_residence_counts.columns = ["EmployeeResidence", "Count"]

# Get the top 10 employee residences
top_10_employee_residences = employee_residence_counts.head(10)

top_10_employee_residences['EmployeeResidenceFullName'] = top_10_employee_residences['EmployeeResidence'].apply(get_country_name)

fig = px.bar(top_10_employee_residences, x="EmployeeResidenceFullName", y="Count", title="Top 10 Employee Residence Distribution")
fig.update_layout(yaxis_type="log") # Change y-axis to logarithmic scale
fig.update_xaxes(title="Employee Residence") # Update x-axis title
fig.show()
```
